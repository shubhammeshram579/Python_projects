## Basics of Descriptive Statistics For Data Science

Basics of Statistics encompasses essential concepts for handling data. It involves data collection, summarization, and interpretation. Descriptive statistics help in understanding data through measures like mean, median, and mode. Variability is assessed using measures like variance and standard deviation.

Inferential statistics draw conclusions and make predictions about populations based on sample data. Key tools include hypothesis testing and confidence intervals. Probability theory underlies statistical methods, enabling the quantification of uncertainty.

Statistics is vital in research, business, healthcare, and more, aiding data-driven decision-making. Understanding distributions, correlations, and regression is essential for modeling relationships. Visualization techniques, like histograms and scatter plots, aid in data exploration. Mastery of statistical fundamentals empowers individuals to extract valuable insights from data.

## Measures of central tendency
Measures of central tendency in statistics, like the mean, median, and mode, summarize the "typical" or central value in a dataset. They help provide a single representative value that describes the central position of the data, aiding in understanding its overall characteristics and making it easier to analyze and compare datasets.

### mean
The mean is the average of a set of values, calculated by summing them and dividing by the count.

```python
import statistics as st
# find the avarage of the Age in this group
Age_mean = [12,18,24,30,33,45,60,34,75,34,65]
print(st.mean(Age_mean))
```

### median
The median is the middle value in a dataset when arranged in ascending order, separating the higher and lower halves.

```python
# find the middle of the Age in this group
Age_median = [12,18,24,30,33,45,60,34,75,34,65,99]
print(st.median(Age_median))
```

### mode
The mode is the value that appears most frequently in a dataset, representing the highest occurrence.

```python
# find the most frequently of the Age in this group
Age_mode =[12,18,24,30,33,45,60,34,75,34,65,125]
print(st.mode(Age_mode))
```

## Five-Number of summary
A five-number summary in statistics provides a concise overview of a dataset: minimum, first quartile, median, third quartile, and maximum. These values give insight into data's range, spread, and central tendency.

```python
#Box plots visually display the five-number summary of a dataset efficiently.
import seaborn as sns

sns.boxplot(Age_mode)
```

### Observation:
In the given box plot, the lower quartile (Q1(25%)) is 28, the median is 34, and the upper quartile (Q3(75%)) is 61. Any data point greater than 120 is considered an outlier.

```python
# find the first quartile and therd quartile value 
q1,q3 =np.percentile(Age_mode,[25,75])
```

```python
print(q1)
```

```python
print(q3)
```

```python
# Range between Q1 and Q3 contains the middle 50% of data.
IQR = q3-q1
lower_fence = q1-1.5*(IQR)
higher_fence=q3+1.5*(IQR)
print(lower_fence,higher_fence)
```

## Measure of Dispersion
A measure of dispersion in statistics quantifies the spread or variability of data points within a dataset. Common measures include the range, which is the difference between the maximum and minimum values, and the standard deviation, which indicates how data points deviate from the mean, offering insights into data distribution and variability. These measures help assess the consistency and reliability of data and are essential for making informed decisions and drawing meaningful conclusions in statistical analysis.

1. Variance
2. Standarsd Devision

```python
#statistics.variance calculates the sample variance in Python's statistics module. 
# It's primarily used for simple statistical calculations.
st.variance(Age_mode)
```

```python
# Manually calculate sample variance

def variance(data):
    n=len(Age_mode)
    mean=sum(data)/n # Mean of the data
    deviation=[(x - mean) ** 2 for x in data] #Varience
    variance=sum(deviation)/(n-1)
    return variance
```

```python
variance(Age_mode)
```

```python
# numpy.var, from the NumPy library, computes both sample and population variances. 
# NumPy is more versatile, offering additional options for calculation methods and axis selection in multidimensional arrays.
np.var(Age_mode,axis=0)
```

```python
# Manually calculate Population variance

def variance(data,dof=0):
    n=len(Age_mode)
    mean=sum(data)/n # Mean of the data
    deviation=[(x - mean) ** 2 for x in data] #Varience
    variance=sum(deviation)/n
    return variance
```

```python
variance(Age_mode)
```

## Gaussian distribution, normal distribution, Log-normal distribution
"Gaussian distribution," often referred to as the "normal distribution," is a symmetrical probability distribution commonly used to model data in various fields. It has a bell-shaped curve with mean and standard deviation parameters. "Log-normal distribution" describes data whose logarithms follow a normal distribution, often used for positive-valued variables like income or stock prices. "Gadias distribution" isn't a standard term in statistics or probability theory; it might be a specific or less common distribution not widely recognized.

### Using Normal Probability Plot to Check Normality
The Normal Probability Plot is a graph that allows us to assess whether or not the data comes from a normal distribution. Note! This plot should be used as a guide for us to assess if the assumption that the data come from a normal distribution is valid or not.

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize=(15,10))
    plt.subplot(2,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(2,2,2)
    stat.probplot(sample,dist="norm",plot=pylab)
    plt.show()
```

```python
plot_data(df["sepal_width"])
```

### observation:
the sepal width data using probability plots, the data points appear to align in a way that resembles a normal distribution.






```python
# creating manually normal distribution data using numpy
a = np.random.normal(1,5,1000)
sns.histplot(a,kde=True)
```

```python
plot_data(a)
```

#### observation:
The data points appear to align in a way that resembles a normal distribution.

## 2. Log-normal distribution 

```python
# creating manually  log normal distribution data using numpy
b = np.random.lognormal(3.,1.,200)
sns.histplot(b,kde=True)
```

```python
def plot_data_log(sample):
    plt.figure(figsize=(15,10))
    plt.subplot(2,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(2,2,2)
    stat.probplot(np.log(sample),dist="norm",plot=pylab)
    plt.show()
```

```python
plot_data_log(b)
```

#### observation:
The data points appear to align in a way that resembles a log normal distribution.

```python
#To convert a log-normal distribution to a normal distribution using NumPy:np.log(data)
# where data is log-normally distributed.
sns.histplot(np.log(b),kde=True)
```
